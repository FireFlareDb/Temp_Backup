Q2. Given  a  decimal  number  as  input,  write  a  program  to  convert  the  given  decimal number into equivalent binary number.

```python
decimal = int(input("Decimal Value: "))
binary = ''
while (decimal > 0):
  q = decimal // 2
  r = decimal % 2
  binary = str(r) + binary
  decimal = q
print("\nBinary: ", binary)
```

Q3. Write  an  Algorithmand  draw  a  flowchart to  Calculate  GCD  and  LCM  of  three numbers.If user enters more than three numbers appropriate message should be displayed.

```python
# Functions
def gcd(a, b):
    """
    This is return gcd of a, b
    """
    while b:
        a, b = b, a % b
    return a

def lcm(a, b):
    """
    This is return lcm of a, b
    
    dependency
    ----------
    To run this fucntion properly you must have gcd function predefined
    """

    _gcd = gcd(a, b)
    lcm = a * b // _gcd
    return lcm

# Inputs
print("Enter Three , Separated Values")
input_list = list(map(int, (input(": ").split(","))))

# Display Answer On Validation
if len(input_list) > 3 or len(input_list) != 3:
    print("HAHA! No Cheating Get Out From Here {Rasta Waha Hai} ğŸ‘‰ğŸ‘‰ ğŸšª")
else:
    a = input_list[0]
    b = input_list[1]
    c = input_list[2]

    print("GCD:",gcd(gcd(a, b), c))
    print("LCM:",lcm(lcm(a, b), c))
```

Q6. Design an algorithm and draw a flowchart which generates even numbers between 1000 and 2000 and then prints them in the standard output. It should also print total sum.

```python
def even(s_range=1000, e_range=2001):
    yield [i for i in range(s_range, e_range) if i % 2 == 0]
    
# Proof Of Concept - 1
print(even())

print()

# POC - 2
for i in even():
    print(i)
    _sum = sum(i)

print("\nSum of all even numbers", _sum)
```

Q9. Design an algorithm and draw a flowchart for the problem of determining whether the number is Armstrong or not.

```python
num = input("n: ")
lenght = len(num)
_sum: int = 0
for i in num:
    result = int(i) ** lenght
    _sum = result + _sum

if _sum == int(num):
    print(f"{num} Is A Armstrong Number")
else:
    print(f"{num} Is Not A Armstrong Number")
    
# Python: Yah... {Humblling} I Can Do This In One Line As Well
```

Q11. Write an algorithm and draw a flowchart to find the factorial of odd numbers from 1 to 100.
