```python
import numpy as np
from typing import Tuple

def modulate_key_n_secret(key: str, secret: str) -> Tuple[np.ndarray, np.ndarray]:
  key = key.replace(" ", "0")
  lenght_key = len(key)
  square_matrix_d = int(np.ceil(len(key) ** 0.5))
  square_matrix_ele = square_matrix_d * square_matrix_d
  if lenght_key != square_matrix_ele:
    differ = int(square_matrix_ele - lenght_key)
    key = key + "0" * differ
  keyACSII = [ord(i) for i in key]
  keyarr = np.array(keyACSII)
  final_key_mat = keyarr.reshape(square_matrix_d, square_matrix_d)
  
  secret = secret.replace(" ", "0")
  lenght_secret = len(secret)
  if lenght_secret != square_matrix_ele:
    while len(secret) % square_matrix_d != 0:
      secret = secret + "0"
  secretACSII = [ord(i) for i in secret]
  secretarr = np.array(secretACSII)
  final_secret_mat = secretarr.reshape(square_matrix_d, -1)

  return final_key_mat, final_secret_mat

def encrypt(keyarr: np.ndarray, secretarr: np.ndarray) -> np.ndarray:
  return np.dot(keyarr, secretarr)

def decrypt(keyarr: np.ndarray, encryptarr: np.ndarray) -> None:
  keyinv = np.linalg.inv(keyarr)
  decoded_mat = np.dot(keyinv, encryptarr)
  new_decoded = decoded_mat.reshape(1, -1)[0]
  message = []
  for i in new_decoded:
    message.append(chr(round(i)))
  print("".join(message).replace("0", " "))


if __name__ == "__main__":
  key: str = "mekeyhu"
  secret: str = "This is my secret"
  keyarr, secretarr = modulate_key_n_secret(key, secret)
  encrypted_arr = encrypt(keyarr, secretarr)
  decrypt(keyarr, encrypted_arr)
```
