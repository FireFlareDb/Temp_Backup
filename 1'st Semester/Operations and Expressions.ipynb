Q2. Find whether a number y is a factor of number x.

```python
x = int(input("X: "))
y = int(input("Y: "))
for i in range(1, x + 1):
    if x % i == 0 and i == y:
        print(f"Yes {y} Is A Factor Of {x}")
        break
else:
    print(f"{y} Is Not A Factor Of {x}")
```

Q5. A day has 86,400 secs (24x60x60). Given a number in the range 1 to 86,400, output the current time as hours, minutes, and seconds with a 24-hour clock. For example:70,000 sec is 19 hours, 26 minutes, and 40 seconds.

```python
time_sec = int(input("Seconds: "))
hour = time_sec // 3600
remain_time_sec = time_sec % 3600
minute = remain_time_sec // 60
remain_time_sec = time_sec % 60
second = remain_time_sec
print(f"Time:\t  {hour}:{minute}:{second}")
```

Q6. A telephone directory has N lines on each page and each page has exactly C columns. An entry in any column has a name with the corresponding telephone number. On
which page, column, and line is the Xth entry (name and number) present? (Assume that page, line, column numbers, and X all start from 1.)

```python
p = int(input("Enter the number of pages: "))
c = int(input("Enter the number of columns: "))
r = int(input("Enter the number of rows: "))
find = int(input("Which entry do you want: "))
total = p*r*c

if find <= total:
  col = find % c
if col == 0:
  col += c

if find % (c*r) == 0:
  page = find / (c*r)
else:
  page = find // (c*r) + 1

row = 0
while find > 0:
  row += 1
  find = find - c

if row > r:
  row = row - (r*(page-1))
  print(f"Go to page number: {page}, row: {row}, column: {col}")
else:
  print("Enter valid values")
```

Q7. If  the  lengths  of  the  two  parallel  sides  of  a  trapezoid  are X meters  and Y meters, respectively,  and  the  height  is H meters,  what  is  the  area  of  the  trapezoid? Write Python code to output the area.

```python
X = float(input("X: "))
Y = float(input("Y: "))
H = float(input("H: "))

area_trapezoid = (1/2) * (X + Y) * H
print("Area Of Trapezoid", area_trapezoid)
```

Q9. Body  massindex (BMI) is a number calculated from a personâ€™s weight and height. According  to  the  Centers  for  Disease  Control  and  Prevention,  the  BMI  is  a  fairly reliable  indicator  of  body  fatness  for  most  people.  BMI  does  not  measure  body  fat directly, but research has shown that BMI correlates to direct measures of body fat, such as underwater weighing and dual-energy X-ray absorptiometry. The formula for BMI is weight/height where weight is in kilograms and height is in meters. 
- Write  a program  that  prompts  for  metric  weight  and  height  and  outputs  the BMI. 
- Write  a  program  that  prompts  for  weight  in  pounds  and  height  in  inches, converts the values to metric, and then calculates the BMI.

```python
def normalizer(pounds: int, inches: int) -> tuple:
    return pounds * 0.4536, inches * 0.0254
    
def BIM(weight: float, height: float) -> int:
    bim = int(weight / height ** 2)
    return bim
```

```python
# First
weight = float(input("Weight: "))
height = float(input("Height: "))
print("Your Body Mass Index", BIM(weight, height))
```

```python
# Second
weight = float(input("Weight In Pounds: "))
height = float(input("Height In Inches: "))
w, h = normalizer(weight, height)
print("Your Body Mass Index", BIM(w, h))
```
