Question 1:

Write a program to calculate grade of student.
0-29: F

30-39:D

40:49:C

50:59:B

60-69:B+

70-79:A

80-89:A+

90-100:O

.......

Take marks of 3 subjects (out of 50 each and calculate above grades)

```python
math, science, sst = int(input("Math: ")), int(input("Science: ")), int(input("SST: "))
total = math + science + sst
percentage = (total / 150) * 100
print("Percentage:", percentage, "%")

if percentage <= 29:
    print("GREAD: F")
elif percentage <= 39:
    print("GREAD: D")
elif percentage <= 49:
    print("GREAD: C")
elif percentage <= 59:
    print("GREAD: B")
elif percentage <= 69:
    print("GREAD: B+")
elif percentage <= 79:
    print("GREAD: A")
elif percentage <= 89:
    print("GREAD: A+")
else:
    percentage <= 100
    print("GREAD: O")
```

```python
lst = [i for i in range(1, 32)]
to_remove = []
counter = 0

while counter != 5:
  val = int(input())
  if val not in to_remove:
    to_remove.append(val)
    counter += 1
  else:
    print("\nAlready Entered, Try Again !!!")

for i in to_remove:
  if i in lst:
    lst.remove(i)

print(lst)
```
