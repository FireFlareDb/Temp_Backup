```python
# marks = data.loc[:, "Quiz":]
# total = marks.sum(axis=1)
# data["Total"] = total
data['Total']=data['Quiz']+ data['Mid Term-I']+ data['Mid Term-II'] + round(data['Pre-Comprehensive']*40/60,2) 
data
```

```python
def grade_me(mark):
    # out_of = 120
    if mark >= 80:
        grade = "A"
    elif mark >= 65:
        grade = "B"
    elif mark >= 45:
        grade = "C"
    else:
        grade = "D"
    return grade

grades = data["Total"].apply(lambda mark: grade_me(mark))
data["Grade"] = grades
data
```
