```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("erdata.csv")
data
```

```python
data.drop(["Zone_Id", "booking_method"], axis=1, inplace=True)
data
```

```python
entry_hour = data["entry_time"].apply(lambda x: int(x[11:13]))
exit_hour = data["exit_time"].apply(lambda x: int(x[11:13]) + 1 if x[11:16].split(":")[1] != "00" else x[11:13])
data["entry_hour"] = entry_hour
data["exit_hour"] = exit_hour
data
```

```python
da = {"14-15": 0, "15-16": 0, "16-17": 0, "17-18": 0}
dal = ["14-15", "15-16", "16-17", "17-18"]
entry_hour = data["entry_hour"]
exit_hour = data["exit_hour"]

for mat in dal:
    ent, exi = map(int, mat.split("-"))
    count = 0
    for i, j in zip(entry_hour, exit_hour):
        if j >= exi and i <= ent:
            count += 1
    da[mat] = count
    
final = pd.DataFrame()
final["Time Interval"] = da.keys()
final.insert(0, "Data", data["entry_time"][0].split(" ")[0])
final["Vehicles In Parking Lot"] = da.values()
final
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x="Time Interval", y="Vehicles In Parking Lot", data=final)
plt.show()
```
