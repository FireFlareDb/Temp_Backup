Q7. Find The Sum Of The Series 8+88+888+8888+.. n Terms. Take Then From The User.

````python
n = int(input("N: "))
bk = n
res = 0
_sum = 0
for i in range(n):
    res = res + bk
    bk = bk * 10
    _sum += res
print("Sum Of Series", _sum)
````

Q10. Write a program that checks to see if a number N is prime.  A simple approach checks all numbers from 2 up to N, but after some point numbers are checked that need not be checked. For example, numbers greater than âˆšN need not be checked. Write a program  that checks   for primarily and avoids those unnecessary checks.

````python
def is_prime(n: int) -> bool:
    # If n Is 2 Then Return True Mean Prime No., This Will Gonna Help To Manage For Loop
    if n == 2:
        return True
    # Not Prime If Passed Argument n Is Below 1 Or Equal To 1
    if n <= 1:
        return False
    # Even Numbers Are Not Prime So Remove Them Except 2
    if n > 2 and n % 2 == 0:
        return False
    
    max_div = math.floor(math.sqrt(n))
    for i in range(3, max_div + 1, 2):
        if n % i == 0:
            return False
    return True
    
n = int(input("Number: "))
if is_prime(n):
    print(f"{n} Is A Prime Number")
else:
    print(f"{n} Is A Not Prime Number")
````

Q14. Write a  program  which repeatedly takes input of  numbers  from  the user  until enters â€œnotmoreâ€. Once â€œnotmoreâ€ is entered, print out the total, count, and average of the numbers. If anything else is given input, ignore that and continue with taking input.

````python
total: int = 0
count: int = 0
while True:
    num = input("-> ")
    if num == "notmore":
        break
    elif num.isdigit():
        total = total + int(num)
        count += 1
    else:
        pass

print("\nTotal", total)
print("Count", count)
print("Average", total / count)
````

Q15. Mr. Anant Asankhya is the  manager  at  the  INFINITE hotel. The  hotel  has  an infinite amount  of rooms.    One    fine    day,  a finite number    of    tourists    come    to    stay    at    the    hotel. The tourists consist of: â†’ A Captain. â†’ An unknown group of families consisting of members per group where Kâ‰ 1 . The Captain was given a separate room, and the rest were given one room per group. Mr. Anant has an  unordered  list  of randomly  arranged room entries.  The list consists of the room numbers for all of the tourists. The room numbers will appear  K times  per  group except for the Captain's room. Mr. Anant needs you to help him find the Captain's room number. The total number of tourists or the  total  number  of  groups  of  families  is  not  known  to  you. You only know the value of K and the room number list. Find out the Captain's room number.

````python
rooms = list(map(int, input().split()))
members = set()
captain = set()

for room in rooms:
    if room not in members:
        members.add(room)
        captain.add(room)
    else:
        captain.discard(room)
        
print(list(captain)[0])
````
